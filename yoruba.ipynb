```python
# STEP 3: Import libraries and load model
import spacy

# Load the transformer-based NER pipeline (uses XLM-RoBERTa under the hood)
nlp = spacy.load("en_core_web_trf")

# STEP 4: Code-mixed Yoruba-English examples
sentences = [
    "Ngozi went to Obafemi Awolowo University yesterday.",
    "Mo ti ri Governor Sanwo-Olu ni Ikeja Stadium.",
    "Zainab works at Dangote Group in Lagos.",
    "Baba mi wa ni Ibadan, close to Cocoa House.",
    "I saw Sola and Adebayo at the GTBank event."
]

# STEP 5: Run NER on each sentence
for sentence in sentences:
    doc = nlp(sentence)
    print(f"\nðŸ“Œ Sentence: {sentence}")
    for ent in doc.ents:
        print(f"  ðŸ”¹ {ent.text} ({ent.label_})")
```
